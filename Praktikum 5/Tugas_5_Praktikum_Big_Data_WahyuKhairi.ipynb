#**Tugas 5 Praktikum Big Data**

# **Library dan baca dataset**

1. Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Read dataset

```python
url = 'https://raw.githubusercontent.com/WahyuKhairi06/BigData_2311531009_Wahyu-Khairi/refs/heads/main/Praktikum%205/heart_cleaned.csv'
df = pd.read_csv(url)
display(df.head())
```

# **Analisis & Visualisasi Data**


**1. Distribusi Numerik**

Fitur Age (usia)

```python
plt.figure(figsize=(12, 4))

# Subplot 1: Histogram usia
plt.subplot(1, 2, 1)
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Usia Pasien')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')

# Subplot 2: Boxplot usia
plt.subplot(1, 2, 2)
sns.boxplot(x=df['age'], color='lightcoral')
plt.title('Boxplot Usia Pasien')
plt.xlabel('Usia')

plt.tight_layout()
plt.show()
```

histogram fitur chol(kolesterol)

```python
plt.figure(figsize=(12, 4))

# Subplot 1: Histogram kolesterol
plt.subplot(1, 2, 1)
sns.histplot(df['chol'], bins=20, kde=True)
plt.title('Distribusi Kolesterol Serum')
plt.xlabel('Kolesterol (mg/dl)')
plt.ylabel('Frekuensi')

# Subplot 2: Boxplot kolesterol
plt.subplot(1, 2, 2)
sns.boxplot(x=df['chol'])
plt.title('Boxplot Kolesterol Serum')
plt.xlabel('Kolesterol (mg/dl)')

plt.tight_layout()
plt.show()
```

histogram fitur trestbps dan boxplot

```python

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['trestbps'], bins=20, kde=True)
plt.title('Distribusi Tekanan Darah Istirahat (trestbps)')
plt.xlabel('Tekanan Darah Istirahat')
plt.ylabel('Frekuensi')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['trestbps'])
plt.title('Boxplot Tekanan Darah Istirahat (trestbps)')
plt.xlabel('Tekanan Darah Istirahat')

plt.tight_layout()
plt.show()
```

**2. Distribusi Kategorikal**


countplot sex dan pie chart sex

```python
plt.figure(figsize=(12, 4))

# Subplot 1: Countplot jenis kelamin
plt.subplot(1, 2, 1)
sns.countplot(x=df['sex'])
plt.title('Countplot Distribusi Jenis Kelamin')
plt.xlabel('Jenis Kelamin (0=Perempuan, 1=Laki-laki)')
plt.ylabel('Jumlah')

# Subplot 2: Pie chart jenis kelamin
plt.subplot(1, 2, 2)
sex_counts = df['sex'].value_counts()
plt.pie(sex_counts, labels=['Laki-laki (1)', 'Perempuan (0)'], autopct='%1.1f%%', startangle=90)
plt.title('Proporsi Jenis Kelamin')

plt.tight_layout()
plt.show()
```

countplot cp dan piechart

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.countplot(x=df['cp'])
plt.title('Countplot Tipe Nyeri Dada (cp)')
plt.xlabel('Tipe Nyeri Dada')
plt.ylabel('Jumlah')

plt.subplot(1, 2, 2)
cp_counts = df['cp'].value_counts()
plt.pie(cp_counts, labels=cp_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribusi Tipe Nyeri Dada (cp)')

plt.tight_layout()
plt.show()
```

countplot thal dan pie chart

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.countplot(x=df['thal'])
plt.title('Countplot Distribusi Thal')
plt.xlabel('Thal')
plt.ylabel('Jumlah')

plt.subplot(1, 2, 2)
thal_counts = df['thal'].value_counts()
plt.pie(thal_counts, labels=thal_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribusi Thal')

plt.tight_layout()
plt.show()
```

**3. Korelasi Antar Fitur**

```python
correlation_matrix = df.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Fitur Numerik')
plt.show()
```

**5. Pairplot**


```python
sns.pairplot(df[['age', 'chol', 'thalach', 'oldpeak']])
plt.suptitle('Pairplot of Selected Numeric Features', y=1.02)
plt.show()
```

**6. EDA Target-Oriented**


```python
sns.histplot(data=df, x='age', hue='target', bins=20, kde=True)
plt.title('Distribusi Usia Pasien Berdasarkan Target')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()
```
