# **Import Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

pandas = untuk mengelola data
matplotlib dan seaborn = untuk visualisasi
df.head untuk melihat 5 baris pertama

```python
df = pd.read_csv("/content/heart_cleaned.csv")
df.head()
```

# **Visualisasi Distribusi Numerik**

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribusi usia pasien')
plt.xlabel('usia')
plt.ylabel('frekuensi')
plt.show()
```

hisplot = membuat histogram
bins 20 = membagi data menjadi 20 interval
kde = true --> menambahkan kurva distribusi

Boxplot

```python
sns.boxplot(x=df['chol'])
plt.title('Boxplot Kolestrol serum')
plt.show()
```

visualisasi distribusi fitur kategorikal

```python
sns.countplot(x="sex", data=df)
plt.title('Distribusi Jenis Kelamin Pasien')
plt.show()
```

```python
sns.countplot(x="arget", data=df)
plt.title('Distribusi Target Penyakit Jantung')
plt.xlabel('target (0 = Tidak, 1 = ya)')
plt.ylabel('Jumlah')
plt.show()
```

korelasi  antar variabel numerik

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap korelasi antar fitur')
plt.show()
```

# **LATIHAN**

1. histogram distribusi trestbps

```python
sns.histplot(df['trestbps'], bins=20, kde=True)
plt.title('Distribusi tekanan darah pasien')
plt.xlabel('tekanan darah')
plt.ylabel('frekuensi')
plt.show()
```

2. boxplot oldpeak

```python
sns.boxplot(x=df['oldpeak'])
plt.title('Boxplot oldpeak pasien')
plt.show()
```

3. countplot untuk cp

```python
sns.countplot(x="cp", data=df)
plt.title('Distribusi chest pain type')
plt.show()
```

4. pie chart untuk thal

```python
thal_counts = df['thal'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(thal_counts, labels=thal_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Thalassemia')
plt.show()
```

5. korelasi antara age dan thalach

```python
corr_age_thalach = df[['age', 'thalach']].corr()
display(corr_age_thalach)
```

```python
sns.scatterplot(x='age', y='thalach', data=df)
plt.title('Scatter Plot Usia vs. Thalach')
plt.xlabel('Usia')
plt.ylabel('Thalach')
plt.show()
```

6. buat heatmap hanya untuk kolom numerik

```python
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antar Fitur Numerik')
plt.show()
```
