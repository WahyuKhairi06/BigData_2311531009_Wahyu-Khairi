# **IMPORT LIBRARY HEART DAN TAMPILKAN 10 BARIS TERATAS**

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df.head(10)
```

```python
df.shape
```

```python
df.info()
df.describe()
```

```python
#menampilkan data invalid
print("Data dengan ca=4 atau thal=0:")
print(df[(df['ca'] == 4) | (df['thal'] == 0)])

#hapus data invalid
df = df[(df['ca'] != 4) & (df['thal'] != 0)]
```

```python
#jumlah data yang
initial_rows = 303
rows_after_cleaning = len(df)
removed_rows = initial_rows - rows_after_cleaning
percentage_removed = (removed_rows / initial_rows) * 100

print(f"Jumlah data setelah pembersihan: {rows_after_cleaning}")
print(f"jumlah baris yang dihapus : {removed_rows} baris, Persentase data yang dihapus:  {percentage_removed:.2f}%")
```

```python
#pisahkan fitur (X) dan label (Y)
X = df.drop('target', axis=1)
y = df['target']

print ("ukuran X : ", X.shape)
print ("ukuran y : ", y.shape)
```

```python
# Simpan data yang sudah dibersihkan
df.to_csv('heart_cleaned.csv', index=False)

# Simpan data yang sudah diproses (setelah encoding)
X.to_csv('X_preprocessed.csv', index=False)
y.to_csv('y_preprocessed.csv', index=False)

print("Data bersih dan data terproses berhasil disimpan")
```

```python
print ("min-max sebelum scaling :")
print ("min : ", X.min().min())
print ("max : ", X.max().max())
```

```python
# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Lakukan scaling pada data fitur (X)
X_scaled = scaler.fit_transform(X)
```

```python
print ("min-max sebelum scaling :")
print ("min : ", X_scaled.min())
print ("max : ", X_scaled.max())
```

```python
#Visualisasi distribusi target
y.value_counts().plot(kind="bar", title="Distribusi Target (Heart Disease)")
plt.xlabel("Target")
plt.ylabel("Jumlah")
plt.show()
```

```python
#Visualisasi distribusi target piechart
y.value_counts().plot(kind="pie", autopct="%1.1f%%", title="Distribusi Target (Heart Disease)")
plt.ylabel("persentase")
plt.show()
```
