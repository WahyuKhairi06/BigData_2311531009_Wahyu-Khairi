```python
import pandas as pd

df = pd.read_csv('heart.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
df.describe()
```

```python
print("Data dengan ca=4 atau thal=0:")
print(df[(df['ca'] == 4) | (df['thal'] == 0)])

df = df[(df['ca'] != 4) & (df['thal'] != 0)]
print("Jumlah data setelah dibersihkan: ", len(df))
```

```python
#one-hot encoding pada kolom kategorika;
categorical_cols = ['sex', 'cp', 'restecg', 'exang', 'slope', 'thal', 'fbs']

df_encoded = pd.get_dummies (df, columns = categorical_cols, drop_first=True)

print ("Jumlah data setelah one-hot encoding: ", len (df_encoded.columns))
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

#kolom numerik
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

#standarisasi
scaler = StandardScaler()
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

df_encoded[numeric_cols].head()
```

```python
#pisahkan x dan y
X = df_encoded.drop('target', axis=1)
y = df_encoded['target']

print ("ukuran X : ", X.shape)
print ("ukuran y : ", y.shape)
```

```python
#simpan data bersih sebelum transformasi
df.to_csv('heart_clean.csv', index=False)

#simpan data setelah transformasi
df_encoded.to_csv('heart_encoding.csv', index=False)

print ("Data bersih dan data terproses berhasil disimpan")
```
